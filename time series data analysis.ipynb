```python
gold.year = pd.to_datetime(gold.year)
gold.head()
```

```python
gold['month'] = gold['year'].dt.month_name(locale = 'English')
gold['year2'] = gold['year'].dt.year
gold.head(20)
#dates.dt.month_name(locale = 'English')
```

```python
gold = gold.iloc[::-1]
gold.reset_index(inplace = True, drop = True)
```

```python
gold['Gold'].plot(kind = 'kde')
plt.show()
```

```python
# plotting moving average(rolling mean)

price_mean = gold['Gold'].rolling(2, win_type=None).mean()
price_mean.plot(style= ['-','--',':'])
plt.show()
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

model = SimpleExpSmoothing(gold['Gold']).fit(smoothing_level=0.2, optimized = True)

gold['Gold'].plot(marker = 'o', color = 'red')
model.fittedvalues.plot(marker = 'o',color = 'blue')
plt.show()
```

```python
# Lag plot - used to check if series is random or not
from pandas.plotting import lag_plot

lag_plot(gold['Gold'])
plt.show()
# pattern 
```

```python
# exhibits a linear pattern
# this shows that the data are strongly non-random and further suggests that an auto regressive model
# might be appropriate
```

```python
# Partial autocorrelation

from pandas import Series
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf

plot_pacf(gold['Gold'])
plt.show()

plot_acf(gold['Gold'])
plt.show()
```
